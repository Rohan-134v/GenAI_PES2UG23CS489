```python
from dotenv import load_dotenv
load_dotenv()

import getpass
import os
from groq import Groq

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API Key: ")
    with open(".env", "w") as f:
        f.write(f"GROQ_API_KEY={os.environ['GROQ_API_KEY']}\n")

client = Groq(api_key=os.environ["GROQ_API_KEY"])
MODEL_NAME = "llama-3.3-70b-versatile"
```

```python
MODEL_CONFIG = {
    "technical": "You are a highly skilled Technical Support Expert. Provide rigorous, code-focused, and precise solutions. Use proper markdown for code.",
    "billing": "You are an empathetic Billing Support Expert. Focus on company policies, clear financial explanations, and resolving billing disputes patiently.",
    "general": "You are a helpful, casual, and friendly General Assistant. Chat normally and assist with everyday questions."
}

def fetch_mock_bitcoin_price():
    return "$64,230.00"

print("Expert configurations and tools defined!")
```

```python
def route_prompt(user_input):
    routing_prompt = """
    Analyze the user's input and classify it into exactly ONE of the following categories:
    - technical (for programming, bugs, code, errors)
    - billing (for refunds, subscriptions, payments, charges)
    - tool_use (for current prices, real-time data lookups like Bitcoin price)
    - general (for casual conversation, greetings, or unclear queries)

    Return ONLY the exact category name in lowercase. Do not include any other text, punctuation, or explanation.
    """

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            {"role": "system", "content": routing_prompt},
            {"role": "user", "content": user_input}
        ],
        temperature=0
    )

    category = response.choices[0].message.content.strip().lower()

    valid_categories = ["technical", "billing", "tool_use", "general"]
    if category not in valid_categories:
        category = "general"

    return category

def process_request(user_input):
    """Routes the request, triggers tools or loads system prompts, and generates the final response."""
    print(f"\n--- Processing Input: '{user_input}' ---")

    category = route_prompt(user_input)
    print(f"-> Intent Classified As: [{category.upper()}]")

    if category == "tool_use":
        price = fetch_mock_bitcoin_price()
        return f"[Tool Successfully Triggered] I fetched the latest data: The current price of Bitcoin is {price}."

    system_prompt = MODEL_CONFIG[category]

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_input}
        ],
        temperature=0.7
    )

    return response.choices[0].message.content

print("Router and Orchestrator functions loaded!")
```

```python
test_queries = [
    "My python script is throwing an IndexError on line 5.",
    "I was charged twice for my subscription this month.",
    "Hey there! How is your day going?",
    "What is the current price of Bitcoin?"
]

for query in test_queries:
    answer = process_request(query)
    print(f"-> Expert Response:\n{answer}\n")
    print("-" * 50)
```
