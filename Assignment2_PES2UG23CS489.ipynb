```python
from transformers import pipeline, set_seed, GPT2Tokenizer
```

```python
user_notes = """
- Sick leave
- Out on Monday
- Back on Tuesday
"""

prompt = f"""
Task: Write a polite email to a manager based on these notes:
{user_notes}

Email Draft:
Subject: Sick Leave

Dear Manager,

I am writing to let you know that"""
```

```python
drafter = pipeline('text-generation', model='distilgpt2')
```

```python
outputs = drafter(
    prompt, 
    num_return_sequences=1
)
print(outputs[0]['generated_text'])
```
